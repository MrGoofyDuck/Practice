```python
# duplicated() function
import pandas as pd

data={'Number':['1','2','3','3'],
       'Letter':['A','B','C','A']}

df = pd.DataFrame(data)

dupl=df.duplicated()

print(data)

print(dupl)
```

```python
#drop_duplicates()
import pandas as pd

data={'Number':['1','2','3','3'],
       'Letter':['A','B','C','A']}

dupl=df.drop_duplicates()

print(data)

print(dupl)
```
